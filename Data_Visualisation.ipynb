```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./data/hotel_room.csv')

df.head()
```

```python
# rename the column
df = df.rename(
    columns={
        'Hotel Name\t\t\t\t\t\t\t\t\t':'Hotel Name'
    }
)

# change the data type of price to int
df['Price'] = df['Price'].str.replace(',','').astype(int)

df
```

**Visualization of Categorical data in non graphical way**

```python
df['Rating Description'].value_counts()
```

```python
df['Rating Description'].value_counts(normalize=True)*100
```

**Visualization of numerical data in non graphical way**

```python
df.describe()
```

```python
df['Price'].value_counts(bins=4)
```

```python
# custome bins
bin = (800,3500,5000,7000,9000,12000)

df['Price'].value_counts(bins=bin)
```

```python
df['Price'].value_counts(bins=bin, normalize=True)*100
```

**Visualization of Categorical data graphically**

```python
# bar graph

sns.countplot(data=df, x="Rating Description")
```

```python
plt.figure(figsize=(12,10))

plt.subplot(1,1,1)
sns.countplot(data=df, x="Rating Description", hue="Location")

plt.show()
```

```python
# pie chart

df['Rating Description'].value_counts().plot(kind="pie", autopct="%.2f")
```

Loading dataset from sns liberary

```python
df2 = sns.load_dataset("titanic")

df2
```

```python
sns.countplot(data=df2, x = "survived", hue='sex')
```

```python
df2['class'].value_counts().plot(kind="pie",autopct="%.2f")
```

**Visualization of numerical data graphically**

```python
# histogram 

sns.histplot(data=df2[df2["fare"]<500], x="fare")
```

```python
# distplot

sns.distplot(df2["fare"])
```

```python
bin = (1, 10, 50, 100, 300, 600)

df2["fare"].value_counts(bins=bin).plot(kind="bar")
```

```python
bin = (1, 10, 50, 100, 300, 600)

sns.histplot(data=df2, x="fare", bins=bin, hue="sex")
```

**Visualization of relation between a categorical and numerical data graphically**

```python
sns.barplot(data=df2, x="class", y="age",hue="sex",estimator="mean", palette="hls")
```

```python
sns.barplot(data=df2, x="class", y="age",hue="sex",estimator="sum", palette="hls")
```

```python
sns.boxplot(data=df2, x="class", y="age",hue="sex", palette="hls")
```

```python
sns.violinplot(data=df2, x="class", y="age")
```

**Visualization of relationship b/w two categorical data**

```python
sns.countplot(data=df2, x="survived", hue="class", palette="hls")
```

```python
# heat map

x = pd.crosstab(df2['survived'], df2['class'])

print(x)

sns.heatmap(x, annot=True,fmt=".1f", cmap='coolwarm')
```

**Visualization of relationship b/w two numerical data**

```python
df3 = sns.load_dataset("tips")

df3
```

```python
df3.iloc[:, :2].corr()
```

```python
# we can see total bill and tip has a pos relationship

# scatter plot

sns.scatterplot(x="total_bill", y="tip",hue="sex", data=df3)
```

```python
sns.scatterplot(x="total_bill", y="tip",hue="sex",style="smoker", data=df3)
```

```python
# joint plot

sns.jointplot(x="total_bill", y="tip",data=df3)
```

```python
sns.jointplot(x="total_bill", y="tip",hue="sex",data=df3)
```

```python
# show best fit line

sns.jointplot(x="total_bill", y="tip",kind="reg", data=df3)
```
