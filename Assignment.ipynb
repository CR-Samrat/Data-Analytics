*4. Whats the % of GDP growth of India in Year 1994 (as compare to 1993)?*

```python
df_1 = df[(df['Country Name']=="India") & (df["Year"]>1992) & (df["Year"]<1995)]

df_1
```

```python
sns.barplot(x="Year", y="Value", data=df_1)
```

```python
data_1 = df_1.values

incr = ((data_1[1][-1]-data_1[0][-1])/(data_1[0][-1]))*100

print("GDP growth from year 1993 to 1994 is", incr)
```

*Find GDP Growth of India from 1960 - 2016?*

```python
df_2 = df[df['Country Name']=="India"]

df_2
```

```python
sns.lineplot(data=df_2, x="Year", y="Value", label="GDP Growth of India (1960 - 2016)")
```

```python
data_2 = df_2.values

gdp_1960 = data_2[0][-1]
gdp_2016 = data_2[-1][-1]

growth = ((gdp_2016 - gdp_1960)/(gdp_1960))*100

growth
```

```python
print("GDP growth from year 1960 to 2016 is", round(growth, 2))
```
